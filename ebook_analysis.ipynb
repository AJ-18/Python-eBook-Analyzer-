# Loading the book

```python
with open("miracle_in_the_andes.txt", "r", encoding="utf-8") as file:
    book = file.read()
```

```python
type(book)
```

# How many chapters are there?

### With string methods:

```python
book.count("Chapter")
```

### With regex:

```python
import re
```

```python
# Set a pattern to find what you're looking for
# The '+' indicates there may be more than one digit after the space
pattern = re.compile("Chapter [0-9]+")

# Find all occurences of the pattern in a specific text
findings = re.findall(pattern, book)
len(findings)
```
