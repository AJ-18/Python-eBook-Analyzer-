### With regex:

```python
import re
```

```python
# Set a pattern to find what you're looking for
# The '+' indicates there may be more than one digit after the space
pattern = re.compile("Chapter [0-9]+")

# Find all occurences of the pattern in a specific text
findings = re.findall(pattern, book)
len(findings)
```

# What are the sentences where "love" was used?

```python
# '^' is a negation operator, include all but... (in this case a period '.')
# The period located at the very end is a literal, so asking for everything in between
# but a period and then a period at the very end as well.
# The capital letters with the curly brackets at the front is getting rid of the
# space at the front, asking to match a capital letter and it should only occur once
pattern = re.compile("[A-Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
findings = re.findall(pattern, book)
len(findings)
```

# What are the most used words?

```python
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book.lower())
findings[:5]
```

```python
results = {}
for word in findings:
    if word in results.keys():
        results[word] = results[word] + 1
    else:
        results[word] = 1
```

```python
results_list = [(value, key) for (key, value) in results.items()]
sorted(results_list, reverse=True)
```

# Extract the paragraphs where "love" was used

```python
pattern = re.compile("[^\n]+love[^\n]+")
findings = re.findall(pattern, book)
findings[:2]
```

# Extract the chapter titles

```python
pattern = re.compile("[a-zA-Z ]+\n\n")
findings = re.findall(pattern, book)
findings = [item.strip("\n\n") for item in findings]
findings
```

# Function that finds the occurrence of any word

```python
def word_occurence(word):
    try:
        pattern = re.compile(word)
        findings = re.findall(pattern, book.lower())
        
        results = {}
        for word in findings:
            if word in results.keys():
                results[word] = results[word] + 1
            else:
                results[word] = 1
        results_list = [(value, key) for (key, value) in results.items()]
        return results_list[0][0]
    except IndexError:
        print(f'The book does not contain the word "{word}".')
```

# Call the function

```python
print(word_occurence("pillow"))
```
