# Loading the book

```python
with open("miracle_in_the_andes.txt", "r", encoding="utf-8") as file:
    book = file.read()
```

# The most used words (non-articles)

```python
import re
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book.lower())
findings[:5]
```

```python
results = {}
for word in findings:
    if word in results.keys():
        results[word] = results[word] + 1
    else:
        results[word] = 1
```

```python
results_list = [(value, key) for (key, value) in results.items()]
results_list = sorted(results_list, reverse=True)
results_list[:5]
```

```python
from platform import python_version
python_version()
```

```python
import nltk
from nltk.corpus import stopwords
english_stopwords = stopwords.words("english")
```

```python
filtered_words = []
for count, word in results_list:
    if word not in english_stopwords:
        filtered_words.append((word, count))
```

# Sentiment Analysis: What is the most positive and the most negative chapter?

## An example

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
scores = analyzer.polarity_scores("Hey, look how beautiful the trees are. I love them. I really love them.")
```

```python
if scores["pos"] > scores["neg"]:
    print("It is a positive text")
else:
    print("It is a negative text")
```

```python
analyzer.polarity_scores(book)
```

## Chapter sentiment analysis

```python
import re
pattern = re.compile("Chapter [0-9]+")
# 're' has a method that lets you split the text by the pattern, so by chapter in this case
chapters = re.split(pattern, book)
```

```python
# Exclude the beginning space that's adding 1 to the number of actual chapters
chapters = chapters[1:]
```

```python
for nr, chapter in enumerate(chapters):
    scores = analyzer.polarity_scores(chapter)
    print(nr + 1, scores)
```
